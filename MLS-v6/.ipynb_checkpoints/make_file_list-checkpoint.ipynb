```python
import os
import re
import numpy as np
import pandas as pd
```

```python
INPUT_DIRS = [
    '/datasets/kaggle/mls/train_1',
    '/datasets/kaggle/mls/train_2',
    '/datasets/kaggle/mls/train_3',
    '/datasets/kaggle/mls/new/test_1_new',
    '/datasets/kaggle/mls/new/test_2_new',
    '/datasets/kaggle/mls/new/test_3_new',    
]
```

```python
# get files in directory
def get_files(src_dir):
    f = []
    for (dirpath, dirnames, filenames) in os.walk(src_dir):
        f.extend(filenames)
        break
    return f
```

```python
safe_labels_df = pd.read_csv('input/train_and_test_data_labels_safe.csv', index_col='image')
```

```python
mat_files_d = {}

for d in INPUT_DIRS:
    for f in get_files(d):
        m = re.findall(r'(\d+)_(\d+)(_(\d+))?\.mat$', f)
        
        if f in safe_labels_df.index:
            f_safe = safe_labels_df.ix[f]['safe']
            f_class = safe_labels_df.ix[f]['class']
        else: 
            f_safe = 1
            f_class = m[0][3]
            f_class = -1 if '' == f_class else int(f_class)
        
        f_patient = int(m[0][0])
        
        if f_safe:
            mat_files_d[f] = {
                'file': f,
                'path': d + '/' + f,
                'class': f_class,
                'patient': f_patient
            }
            
print len(mat_files_d)
```

```python
input_df = pd.DataFrame(mat_files_d.values())
input_df = pd.concat([input_df, pd.get_dummies(input_df['patient'], prefix='patient')], axis=1)
input_df.drop('patient', inplace=True, axis=1)
```
