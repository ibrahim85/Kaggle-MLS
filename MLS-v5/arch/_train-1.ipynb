```python
import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm
```

```python
MSGS_DIR = 'out/msgs'

IMG_SHAPE = (96, 2048, 1)

# axis 0 is batch sample index
AXIS_FREQ = 1
AXIS_TIME = 2
AXIS_CH = 3
```

```python
import math

imgs = np.zeros((1000,) + IMG_SHAPE, dtype=np.float32)

for e in tqdm(range(len(imgs))):

    vscale = np.random.rand()
    hscale = 1+np.random.rand()

    for i in range(IMG_SHAPE[1]):
        x = float(i) / math.pi
        h = float(IMG_SHAPE[0])
        y = math.sin(x/(10/hscale))*h/2*vscale+h/2
        y = int(y)
        imgs[e][y][i][0] = 1.
        imgs[e][y][i-1][0] = 1.
        imgs[e][y][i-2][0] = 1.
        imgs[e][y][i-3][0] = 1.
        imgs[e][y][i-4][0] = 1.
```

```python
VAL_SIZE = int(0.25 * len(imgs))
X_train = imgs[:-VAL_SIZE]
X_val = imgs[-VAL_SIZE:]
```

```python
img = imgs[0]
from matplotlib import cm
plt.figure(figsize=(10,4))
plt.imshow(img.reshape(IMG_SHAPE[0], IMG_SHAPE[1]), aspect=4, cmap=cm.gray_r)
```

```python
from keras.layers import Input, Dense, Convolution2D, \
    MaxPooling2D, UpSampling2D, BatchNormalization, Dropout, \
    UpSampling2D, Layer
from keras.models import Model, Sequential
from keras import backend as K
from keras.layers.advanced_activations import ELU
from keras.callbacks import TensorBoard
```

```python
model = Sequential()

# conv 1
model.add(Convolution2D(32, 3, 3, border_mode='same', input_shape=IMG_SHAPE))
model.add(ELU())
model.add(MaxPooling2D(pool_size=(2,4)))

# conv 2
model.add(Convolution2D(128, 3, 3, border_mode='same'))
model.add(ELU())
model.add(MaxPooling2D(pool_size=(2,4)))

# conv 3
model.add(Convolution2D(128, 3, 3, border_mode='same'))
model.add(ELU())
model.add(MaxPooling2D(pool_size=(2,4)))

# conv 4
model.add(Convolution2D(192, 3, 3, border_mode='same'))
model.add(ELU())
model.add(MaxPooling2D(pool_size=(2,4)))

# conv 5
model.add(Convolution2D(256, 3, 3, border_mode='same'))
model.add(ELU())
model.add(MaxPooling2D(pool_size=(2,2)))

# deconv 5
model.add(Convolution2D(256, 3, 3, border_mode='same'))
model.add(ELU())
model.add(UpSampling2D(size=(2,2)))

# deconv 4
model.add(Convolution2D(192, 3, 3, border_mode='same'))
model.add(ELU())
model.add(UpSampling2D(size=(2,4)))

# deconv 3
model.add(Convolution2D(128, 3, 3, border_mode='same'))
model.add(ELU())
model.add(UpSampling2D(size=(2,4)))

# deconv 2
model.add(Convolution2D(128, 3, 3, border_mode='same'))
model.add(ELU())
model.add(UpSampling2D(size=(2,4)))

# deconv 1
model.add(Convolution2D(32, 3, 3, border_mode='same'))
model.add(ELU())
model.add(UpSampling2D(size=(2,4)))

# output
model.add(Convolution2D(1, 3, 3, activation='sigmoid', border_mode='same'))
```

```python
model.compile(optimizer='adadelta', loss='binary_crossentropy')
```

```python
model.fit(
    X_train,
    X_train, 
    nb_epoch=30,
    verbose=True,
    validation_data=(X_val, X_val),
    shuffle=False,
    callbacks=[
        TensorBoard(log_dir='/tmp/tf-b')
    ]
)
```
